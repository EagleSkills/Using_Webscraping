# scrap and get links

```python
# import libraries
import requests
from bs4 import BeautifulSoup
```

```python
# create url objects
url = 'https://en.wikipedia.org/wiki/Main_Page'
# determine response as r
r = requests.get(url)
# use Beautiful soup to parse the html page
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
# find the links restricted by start with 'http' only and get the result into list

filtered_links = [a.get('href') for a in soup('a') if a.get('href', ' ').startswith('http')]
filtered_links
  
  
```

```python
# alternative
list = []
tag = soup('a')
for i in tag:
    if i.get('href').startswith('http'):
        list.append(i.get('href'))
list
```
