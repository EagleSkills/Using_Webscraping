```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.timesjobs.com/candidate/job-search.html?searchType=Home_Search&from=submit&asKey=OFF&txtKeywords=&cboPresFuncArea=35'
response = requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.text, 'lxml')
```

```python
values = []
data = soup.find_all('li', class_ = 'clearfix job-bx wht-shd-bx')
for row in data:
    job_name = row.h2.text.strip()
    comp_name = row.h3.text.strip()
    job_des = row.find('ul', class_ = 'list-job-dtl clearfix').li.text.strip().replace('.. More Details', ' ').replace('\n',' ').replace('\r',' ')
    skills = row.find('span', class_ = 'srp-skills').text.strip()
    posted = row.find('span', class_ = 'sim-posted').text.strip()
    values.append({'job_name': job_name, 'company_name':comp_name, 'job_descreption': job_des, 'skills': skills, 'job_bosted':posted})
values    
```

```python
df = pd.DataFrame(values)
df
```

```python
df.to_csv('kalabal.csv', index=False)
```
